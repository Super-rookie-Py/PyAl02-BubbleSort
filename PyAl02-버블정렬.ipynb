### 파이썬 알고리즘 공부 버블 정렬

#### 버블정렬의 기본 알고리즘은 바로

#### "옆에 있는 값과 비교해서 더 작은 값을 앞으로 보내면 어떨끼?"

```python
my_list = [1, 10, 5, 8, 7, 6, 4, 3, 2, 9]
```

```python
for i in range(9):
    for j in range(9-i):
        if my_list[j] > my_list[j+1]:
            my_list[j], my_list[j+1] = my_list[j+1], my_list[j]
            print("중간과정: ", my_list)
print(my_list)
        
```

## 버블 정렬의 시간 복잡도

### 비교 연산의 횟수 9 + 8 + 7 + ... + 1
### 데이터의 갯수가 N 개일 때 n-1 까지의 합
### n*(n-1) / 2 = 45번

#### 데이터의 갯수가 무수히 많을 때 나누거나 더한 값은 의미가 없다.

### 처리시간은 O(N * N) 로 표기할 수 있다.

## 버블정렬의 시간 복잡도는 O(N^2)

#### 버블정렬과 선택정렬의 시간복잡도는 같지만 실제 처리시간은
#### 버블정렬의 처리시간이 더 오래걸린다.
#### 선택정렬은 작은값을 찾아서 바꿔주지만
#### 버블정렬은 바로옆값과 바꿔가면서 찾기 때문에 실제 처리시간이 더걸린다.

#### 매번교체 해주기때문에 실제 수행시간은 오래걸림
#### 정렬 알고리즘 중에 가장 오래걸린다.
